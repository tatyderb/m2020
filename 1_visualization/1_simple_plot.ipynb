## Шаг 1. Точки и plot

Построим графики фукнкций $sin(x)$ и $cos(x)$

Будем улучшать график постепенно.

Сначала нужно построить точки Х и У. Можно задать как раньше, а можно через библиотеку numpy

```

````python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # 258 точек от -пи до пи, включая последнюю, вектор Х
CY, SY = np.cos(X), np.sin(X)                      # векторы Y координат для cos и sin

plt.plot(X, CY)
plt.plot(X, SY)

plt.show()

````

````python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # 258 точек от -пи до пи, включая последнюю, вектор Х
CY, SY = np.cos(X), np.sin(X)                      # векторы Y координат для cos и sin

plt.figure(figsize=(10,6), dpi=80)   # размер фигуры 10 к 6, более горизонтальная
plt.plot(X, CY, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, SY, color="red",  linewidth=5, linestyle="-.")

plt.show()

````


## Границы по х и у

Рисовать от хmin до хmax можно командой


```
plt.xlim(xmin, xmax)
```

Можно найти в векторе минимальное и максимальное значение `CY.min()` и `CY.max()`

и выставить границы так:
```
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(CY.min()*1.1, CY.max()*1.1)
```

````python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # 258 точек от -пи до пи, включая последнюю, вектор Х
CY, SY = np.cos(X), np.sin(X)                      # векторы Y координат для cos и sin

plt.figure(figsize=(10,6), dpi=80)   # размер фигуры 10 к 6, более горизонтальная

# диапазон по х и у
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(CY.min()*1.1, CY.max()*1.1)

plt.plot(X, CY, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, SY, color="red",  linewidth=5, linestyle="-.")

plt.show()

````

## Поставим черточки на шкалах (ticks)

Хотим поставить черточки на шкалах и подписать значения не в -3, -2, -1, 0, 1, 2, 3 на оси Х, а в $-{\pi},  -\pi/2, 0, \pi/2, \pi$

```
plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
```

````python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # 258 точек от -пи до пи, включая последнюю, вектор Х
CY, SY = np.cos(X), np.sin(X)                      # векторы Y координат для cos и sin

plt.figure(figsize=(10,6), dpi=80)   # размер фигуры 10 к 6, более горизонтальная

# диапазон по х и у
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(CY.min()*1.1, CY.max()*1.1)

# ставим черточки на осях
plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])


plt.plot(X, CY, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, SY, color="red",  linewidth=5, linestyle="-.")

plt.show()
````


## Подписи хотим красивые, c $\pi$, а не -3.142

Перечислим список подписей **такой же длины**, что и список черточек. Если не совпадут размеры, будет ошибка.

```
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])```

````python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # 258 точек от -пи до пи, включая последнюю, вектор Х
CY, SY = np.cos(X), np.sin(X)                      # векторы Y координат для cos и sin

plt.figure(figsize=(10,6), dpi=80)   # размер фигуры 10 к 6, более горизонтальная

# диапазон по х и у
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(CY.min()*1.1, CY.max()*1.1)

# ставим черточки на осях
plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])

# подписи с pi на оси Х
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.plot(X, CY, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, SY, color="red",  linewidth=5, linestyle="-.")

plt.show()
````

## Рисуем оси в правильном месте, а не сверху и снизу

В математике принято рисовать оси Х и У через 0. Нарисуем их так.

Обычно в ваших графиках это делать не нужно.

Осей на самом деле было нарисовано 4: right, left, top, bottom.

Сделаем оси right и top невидимыми.

Оси left и bottom передвинем в 0.

```
ax = plt.gca()                          # взять оси графика
ax.spines['right'].set_color('none')    # сделать ось right невидимой
ax.spines['top'].set_color('none')      # сделать ось top невидимой
ax.xaxis.set_ticks_position('bottom')   # взять черточки для осих Х с оси bottom
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')     # взять черточки для осих Y с оси left
ax.spines['left'].set_position(('data',0))
```

````python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # 258 точек от -пи до пи, включая последнюю, вектор Х
CY, SY = np.cos(X), np.sin(X)                      # векторы Y координат для cos и sin

plt.figure(figsize=(10,6), dpi=80)   # размер фигуры 10 к 6, более горизонтальная

# диапазон по х и у
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(CY.min()*1.1, CY.max()*1.1)

# ставим черточки на осях
plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])

# подписи с pi на оси Х
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Передвигаем оси так, чтобы они шли через (0,0)
ax = plt.gca()                          # взять оси графика
ax.spines['right'].set_color('none')    # сделать ось right невидимой
ax.spines['top'].set_color('none')      # сделать ось top невидимой
ax.xaxis.set_ticks_position('bottom')   # взять черточки для осих Х с оси bottom
ax.spines['bottom'].set_position(('data',0))  # поставись ось bottom в точку 0
ax.yaxis.set_ticks_position('left')     # взять черточки для осих Y с оси left
ax.spines['left'].set_position(('data',0))    # поставись ось left в точку 0

plt.plot(X, CY, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, SY, color="red",  linewidth=5, linestyle="-.")

plt.show()
````

## Добавим легенду (что значит какая линия)

Я не понимаю, что значит синяя и красная линии. Нужно объяснить.

Подпишем, что синяя линия это косинус, а красная - синус.

Для этого в `plot` добавим аргумент `label` и скажем, что легенду нужно показать.

```
plt.legend(loc='upper left', frameon=False)
```
Получилось:

````python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # 258 точек от -пи до пи, включая последнюю, вектор Х
CY, SY = np.cos(X), np.sin(X)                      # векторы Y координат для cos и sin

plt.figure(figsize=(10,6), dpi=80)   # размер фигуры 10 к 6, более горизонтальная

# диапазон по х и у
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(CY.min()*1.1, CY.max()*1.1)

# ставим черточки на осях
plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])

# подписи с pi на оси Х
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Передвигаем оси так, чтобы они шли через (0,0)
ax = plt.gca()                          # взять оси графика
ax.spines['right'].set_color('none')    # сделать ось right невидимой
ax.spines['top'].set_color('none')      # сделать ось top невидимой
ax.xaxis.set_ticks_position('bottom')   # взять черточки для осих Х с оси bottom
ax.spines['bottom'].set_position(('data',0))  # поставись ось bottom в точку 0
ax.yaxis.set_ticks_position('left')     # взять черточки для осих Y с оси left
ax.spines['left'].set_position(('data',0))    # поставись ось left в точку 0

plt.plot(X, CY, color="blue", linewidth=2.5, linestyle="-", label="косинус")
plt.plot(X, SY, color="red",  linewidth=5, linestyle="-.", label="синус")

# рисуем легенду в левом верхнем углу без рамочки
plt.legend(loc='upper left', frameon=False)

plt.show()
````

## Название графика (title)

О чем этот график, нужно его назвать.

Добавим title. Он добавляется прямо ко всей figure.

Сделем шрифт графика побольше, `fonsize=20`

```
plt.title("Графики синуса и косинуса", fontsize=20)

````python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # 258 точек от -пи до пи, включая последнюю, вектор Х
CY, SY = np.cos(X), np.sin(X)                      # векторы Y координат для cos и sin

plt.figure(figsize=(10,6), dpi=80)   # размер фигуры 10 к 6, более горизонтальная
plt.title("Графики синуса и косинуса", fontsize=20)

# диапазон по х и у
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(CY.min()*1.1, CY.max()*1.1)

# ставим черточки на осях
plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])

# подписи с pi на оси Х
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Передвигаем оси так, чтобы они шли через (0,0)
ax = plt.gca()                          # взять оси графика
ax.spines['right'].set_color('none')    # сделать ось right невидимой
ax.spines['top'].set_color('none')      # сделать ось top невидимой
ax.xaxis.set_ticks_position('bottom')   # взять черточки для осих Х с оси bottom
ax.spines['bottom'].set_position(('data',0))  # поставись ось bottom в точку 0
ax.yaxis.set_ticks_position('left')     # взять черточки для осих Y с оси left
ax.spines['left'].set_position(('data',0))    # поставись ось left в точку 0

plt.plot(X, CY, color="blue", linewidth=2.5, linestyle="-", label="косинус")
plt.plot(X, SY, color="red",  linewidth=5, linestyle="-.", label="синус")

# рисуем легенду в левом верхнем углу без рамочки
plt.legend(loc='upper left', frameon=False)

plt.show()
````
