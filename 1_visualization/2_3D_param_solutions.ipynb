## Вариант 1. Тор

* $x = cos(u)(cos(v)+3)$, 
* $y = sin(u)(cos(v)+3)$, 
* $z = sin(v)$
* где $u \in [-\pi, \pi], v \in [-\pi, \pi]$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(-2*np.pi, 2*np.pi, 100)
v = np.linspace(-np.pi, np.pi, 100)

x = np.outer(np.cos(u), (np.cos(v)+3))
y = np.outer(np.sin(u), (np.cos(v)+3))
z = np.outer(np.ones(np.size(v)), np.sin(v))
ax.plot_surface(x, y, z, rstride=4, cstride=4, color='b')
ax.set_zlim(-3, 3)


plt.show()
fig.savefig('3p_1.png')
```

## Вариант 2. Чаша

* $x = cos(u)(cos(v)+3)$, 
* $y = sin(u)(cos(v)+3)$, 
* $z = sin(v) + u$
* где $u \in [-2\pi, 2\pi], v \in [-\pi, 0]$

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(-2*np.pi, 2*np.pi, 100)
v = np.linspace(-np.pi, 0, 100)

x = 1*np.outer(np.cos(u), (np.cos(v)+3))
y = 1*np.outer(np.sin(u), (np.cos(v)+3))
z = 1*np.outer(np.ones(np.size(v)), np.sin(v)+u)
ax.plot_surface(x, y, z, rstride=4, cstride=4, cmap='jet')
#ax.set_xlim(-3, 6)
#ax.set_zlim(-8, 8)


plt.show()
fig.savefig('3p_2.png')
```
