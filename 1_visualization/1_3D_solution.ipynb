# Вариант 1. 

$z = xy$, где $x \in [-6, 6], y \in [-6, 6]$, заливка 'hot'

```python
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import axes3d   # новый import
fig = plt.figure()
ax = axes3d.Axes3D(fig)          # создаем axes из figure

xi = np.arange(-1, 1, 0.01)               # точки по оси Х
yi = np.arange(-1, 1, 0.01)               # точки по оси Y
X, Y = np.meshgrid(xi, yi)                # X, Y - узлы на сети 
Z = X*Y                           # в каждой точке добавили z координату

ax.plot_surface(X, Y, Z, rstride=10, cstride=10, cmap='hot')   # заливка цветом
ax.set_title('$z=xy$')
plt.show()
fig.savefig('3_1.png')
```

# Вариант 2. 

$z = x^2+y^2$, где $x \in [-6, 6], y \in [-6, 6]$, заливка 'winter'

```python
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import axes3d   # новый import
fig = plt.figure()
ax = axes3d.Axes3D(fig)          # создаем axes из figure

xi = np.arange(-1, 1, 0.01)               # точки по оси Х
yi = np.arange(-1, 1, 0.01)               # точки по оси Y
X, Y = np.meshgrid(xi, yi)                # X, Y - узлы на сети 
Z = X*X +Y*Y                           # в каждой точке добавили z координату

ax.plot_surface(X, Y, Z, rstride=10, cstride=10, cmap='winter')   # заливка цветом
ax.set_title('$z=x^2 + y^2$')
plt.show()
fig.savefig('3_2.png')
```

# Вариант 3 

$z = sin(x)cos(y)$, где $x \in [-2\pi, 2\pi], y \in [-2\pi, 2\pi]$, заливка 'jet'

```python
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import axes3d   # новый import
fig = plt.figure()
ax = axes3d.Axes3D(fig)          # создаем axes из figure

xi = np.arange(-2*np.pi, 2*np.pi, 0.02)               # точки по оси Х
yi = np.arange(-2*np.pi, 2*np.pi, 0.02)               # точки по оси Y
X, Y = np.meshgrid(xi, yi)                # X, Y - узлы на сети 
Z = np.sin(X)*np.cos(Y)                           # в каждой точке добавили z координату

ax.plot_surface(X, Y, Z, rstride=10, cstride=10, cmap='jet')   # заливка цветом
ax.set_title('$z=sin(x)cos(y)$')
plt.show()
fig.savefig('3_3.png')
```

# Вариант 4

$z = exp(-30(x^2 + y^2))$, где $x \in [-1, 1], y \in [-1, 1]$, заливка 'viridis'

```python
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import axes3d   # новый import
fig = plt.figure()
ax = axes3d.Axes3D(fig)          # создаем axes из figure

xi = np.arange(-1, 1, 0.01)               # точки по оси Х
yi = np.arange(-1, 1, 0.01)               # точки по оси Y
X, Y = np.meshgrid(xi, yi)                # X, Y - узлы на сети 
Z = np.exp(-30*(X*X + Y*Y))                           # в каждой точке добавили z координату

ax.plot_surface(X, Y, Z, rstride=10, cstride=10, cmap='viridis')   # заливка цветом
ax.set_title('$exp(-30(x^2 + y^2))$')
plt.show()
fig.savefig('3_4.png')
```

# Вариант 5

$z = sin(x^2 + y^2)$, где $x \in [-\pi/2, \pi/2], y \in [-\pi/2, \pi/2]$, заливка 'viridis'

```python
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import axes3d   # новый import
fig = plt.figure()
ax = axes3d.Axes3D(fig)          # создаем axes из figure
k = 0.5
xi = np.arange(-np.pi*k, np.pi*k, 0.01)               # точки по оси Х
yi = np.arange(-np.pi*k, np.pi*k, 0.01)               # точки по оси Y
X, Y = np.meshgrid(xi, yi)                # X, Y - узлы на сети 
Z = np.sin(X*X + Y*Y)                           # в каждой точке добавили z координату

ax.plot_surface(X, Y, Z, rstride=10, cstride=10, cmap='viridis')   # заливка цветом
ax.set_title('$sin(x^2 + y^2)$')
plt.show()
fig.savefig('3_5.png')
```
